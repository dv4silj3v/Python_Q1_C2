#### Task 2

```python
t = np.linspace(0.0, 10.0, num=51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])

plt.plot(t, f, 'g')

plt.title("Graph f(t)", fontsize=16, fontweight="bold")

plt.xlabel("Variable t")
plt.ylabel("Function f(t)")

plt.show()
```

#### Task 3**

```python
x = np.linspace(-3, 3, num=51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ay1, ay2, ay3, ay4 = ax.flatten()

ay1.plot(x, y1)
ay1.set_title("Graph y1")
ay1.set_xlim([-5, 5])

ay2.plot(x, y2)
ay2.set_title("Graph y2")

ay3.plot(x, y3)
ay3.set_title("Graph y3")

ay4.plot(x, y4)
ay4.set_title("Graph y4")

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
creditcardfile = 'creditcard.csv'
ccfd = pd.read_csv(creditcardfile)
```

```python
plt.style.use('fivethirtyeight')
```

```python
ccfd.shape
```

```python
ccfd.columns
```

```python
ccfd.iloc[0:10,[0, 29, 30]] 
```

```python
ccfd_class_series = ccfd.set_index('Time')['Class']
ccfd_class_series.head()
```

```python
ccfd_class_series.value_counts()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 4, 4
```

```python
ccfd_class_series.value_counts().plot(kind='bar')
plt.show()
```

```python
rcParams['figure.figsize'] = 4, 4
ccfd_class_series.value_counts().plot(kind='bar', logy=True)
plt.show()
```

```python
plt.hist(ccfd[ccfd["Class"] == 1]["V1"], density = 1, bins = 20, alpha = 0.5, color = "red", label = 'Class 1')
plt.hist(ccfd[ccfd["Class"] == 0]["V1"], density = 1, bins = 20, alpha = 0.5, color = "grey", label = 'Class 0')
plt.xlabel("Class")

plt.legend()

plt.show()
```
